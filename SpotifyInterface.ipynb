```python
import spotipy
import json
import webbrowser
import pandas
```

```python
import urllib.request
import spotipy.util as util
```

```python
url = "https://api.open-meteo.com/v1/forecast?latitude=-25&longitude=135&current=weather_code&hourly=temperature_2m,weather_code&daily=weather_code&timezone=GMT"
```

```python
#This is like puuting the url in a web browser
request = urllib.request.Request(url)
```

```python
#Response is what you get back from entering a URL - json, html, status
response = urllib.request.urlopen(request)
```

```python
response.status
```

```python
print(response)
```

```python
#Read and load in all hte json data from the response object
#ie - ignore meta dat that I do not care
weather = json.loads(response.read())
```

```python
weather
```

```python
weather["hourly"]
```

```python
#With dictionaries, I need a key, with lists I need an index
temp = weather["hourly"]["temperature_2m"][0]
```

```python
temp
```

```python
temp_written = ""
if temp >= 25 and temp <= 45:
    temp_written = "hot"
elif temp >= 20 and temp < 25:
    temp_written = "warm"
elif temp >= 15 and temp < 20:
    temp_written = "mild"
elif temp >= 10 and temp < 15:
    temp_written = "cool"
elif temp >= 5 and temp < 10:
    temp_written = "cold"
else:
    temp_written = "freezing"
    
```

```python
print(temp_written)
```

```python
#Open file with keys 
credentials = "spotify-keys.json"
with open(credentials, "r") as keys:
    api_tokens = json.load(keys)
```

```python
api_tokens
```

```python
client_id = api_tokens["client_id"]
client_secret = api_tokens["client_secret"]
redirectURI = api_tokens["redirect"]
username = api_tokens["username"]
```

```python
username
```

```python

scope = 'user-read-private user-read-playback-state user-modify-playback-state playlist-modify-public'

#Getting a permission from spotify
token = util.prompt_for_user_token(username, scope, client_id=client_id,
                           client_secret=client_secret,
                           redirect_uri=redirectURI)
```

```python
token
```

```python
#Create Spotify Object
# Create OAuth Object
# oauth_object = spotipy.SpotifyOAuth(client_id, client_secret, redirectURI)
# Create token
# token_dict = oauth_object.get_cached_token()
# token = token_dict['access_token']
# print(token_dict)
# Create Spotify Object
sp = spotipy.Spotify(auth=token)
```

```python
devices = sp.devices()
print(devices)
deviceID = devices['devices'][0]['id']
```

```python
# learn about more query strings for spotify here: https://developer.spotify.com/documentation/web-api/reference/#/operations/search
track_results = sp.search(q=temp_written, type='track', limit=50)
#track_results = sp.search(q="artist:" + temp_written, type='track', limit=50)
```

```python
tracks = track_results['tracks']['items']
```

```python
track_selection_list = []

for song in tracks:
     track_selection_list.append(song['uri'])
    
print(track_selection_list)
```

```python
my_playlist = sp.user_playlist_create(user=username, name="My awesome playlist", public=True,
                                      description="Songs for the weather")
```

```python
my_playlist
```

```python
results = sp.user_playlist_add_tracks(username, my_playlist['id'], track_selection_list)
print(results)
```
