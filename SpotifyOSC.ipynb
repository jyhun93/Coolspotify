```python
name = json.loads(response.read())
```

```python
Age = name["age"]
print("Predicted age:" + str(Age))
```

```python
age_wriiten = ""
if Age >= 69 and Age <= 95:
    age_written = "1955"
elif Age >= 43 and Age <= 68:
    age_written = "1975"
elif Age >= 25 and Age <= 42:
    age_written = "1995"

age_written
```

```python
#Open file with keys 
credentials = "spotifyOSC.json"
with open(credentials, "r") as keys:
    api_tokens = json.load(keys)
```

```python
api_tokens
```

```python
client_id = api_tokens["client_id"]
client_secret = api_tokens["client_secret"]
redirectURI = api_tokens["redirect"]
username = api_tokens["username"]
```

```python
username
```

```python
scope = 'user-read-private user-read-playback-state user-modify-playback-state playlist-modify-public'
token = util.prompt_for_user_token(username, scope, client_id=client_id,
                           client_secret=client_secret,
                           redirect_uri=redirectURI)
```

```python
token
```

```python
# Create Spotify Object
# Setting the authorization with the final key, ie the token
sp = spotipy.Spotify(auth=token)
```

```python
devices = sp.devices()
print(devices)
deviceID = devices['devices'][0]['id']
```

```python
track_results = sp.search(q="year:" + age_written, type='track', limit=50)
```

```python
tracks = track_results['tracks']['items']
```

```python
track_selection_list = []
track_selection_id = []

for song in tracks:
     track_selection_id.append(song['id'])
     track_selection_list.append(song['uri'])
     
#print(track_selection_id)
```

```python
features = sp.audio_features(track_selection_id)
features
```

```python
tempo_value = 0
energy_value = 0
liveness_value = 0
loudness_value = 0
valence_value = 0

for x in features:
    tempo_value += x['tempo']/50
    energy_value += x['energy']/50
    liveness_value += x['liveness']/50
    loudness_value += x['loudness']/50
    valence_value += x['valence']/50

#tempo_value
print(str(tempo_value) + ", " + str(energy_value) + ", " + str(liveness_value) + ", " + str(loudness_value) + ", " + str(valence_value))
```

```python
exectime = time.time() + 10   # execute in 10 seconds
age = oscbuildparse.OSCMessage("/sound/age", None, [str(Age)])
tempo = oscbuildparse.OSCMessage("/sound/tempo", None, [tempo_value])
loudness = oscbuildparse.OSCMessage("/sound/loudness", None, [loudness_value])
#energy = oscbuildparse.OSCMessage("/sound/energy", None, [energy_value])
#liveness = oscbuildparse.OSCMessage("/sound/liveness", None, [liveness_value])
#valence = oscbuildparse.OSCMessage("/sound/freq", None, [valence_value])

bun = oscbuildparse.OSCBundle(oscbuildparse.unixtime2timetag(exectime), 
                              [age, tempo, loudness])
osc_send(bun, "processing")

#osc_terminate(
```

```python
msg = osc_message_builder.OscMessageBuilder(address="/test")
msg.add_arg(Age)
msg.add_arg(tempo_value)
msg.add_arg(loudness_value)
#msg.add_arg(energy_value)
#msg.add_arg(liveness_value)
#msg.add_arg(valence_value)

msg = msg.build()
client.send(msg)
```
